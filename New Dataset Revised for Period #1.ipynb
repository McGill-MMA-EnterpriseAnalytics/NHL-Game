```python
import pandas as pd
import numpy as np
pd.set_option("display.max_rows", None)
pd.set_option('display.max_columns', None)
```

```python
gd = pd.read_csv("/Users/matthewbuttlerives/Downloads/game.csv")
gp = pd.read_csv("/Users/matthewbuttlerives/Downloads/game_plays.csv")
gts = pd.read_csv("/Users/matthewbuttlerives/Downloads/game_teams_stats.csv")
```

```python
# Selecting data for first and second 
Period_1 = gp[(gp.period == 1)]
```

```python
column_values = Period_1["event"].values.ravel()
unique_values =  pd.unique(column_values)
unique_values
```

```python
column_values = Period_1["secondaryType"].values.ravel()
unique_values =  pd.unique(column_values)
unique_values
```

```python
# Preparing to group data together
test = Period_1[["game_id","team_id_for","event"]]
```

```python
test.shape
```

```python
# Many missing null values do not have a team associated to them such as Period Ready or Period Start
test.isnull().sum()
```

```python
test = test.dropna()
```

```python
test.shape
```

```python
# Creating dataframe of shots on net
test = test[(test.event == "Shot")|
             (test.event == "Goal")|(test.event == "Takeaway")|
             (test.event == "Hit")|(test.event == "Blocked Shot")|
            (test.event == "Giveaway")|(test.event == "Missed Shot")|
                    (test.event == "Penalty")|(test.event == "Faceoff")]
```

```python
Shot_test = test[(test.event == "Shot")]
tq = Shot_test.groupby(["game_id","team_id_for"]).size()
tq = pd.DataFrame(tq)
tq = tq.reset_index()
tq = tq.rename(columns={0: "Shots"})
```

```python
Goal_test = test[(test.event == "Goal")]
tq2 = Goal_test.groupby(["game_id","team_id_for"]).size()
tq2 = pd.DataFrame(tq2)
tq2 = tq2.reset_index()
tq2 = tq2.rename(columns={0: "Goals"})
```

```python
Take_test = test[(test.event == "Takeaway")]
tq3 = Take_test.groupby(["game_id","team_id_for"]).size()
tq3 = pd.DataFrame(tq3)
tq3 = tq3.reset_index()
tq3 = tq3.rename(columns={0: "Takeaways"})
```

```python
Hit_test = test[(test.event == "Hit")]
tq4 = Hit_test.groupby(["game_id","team_id_for"]).size()
tq4 = pd.DataFrame(tq4)
tq4 = tq4.reset_index()
tq4 = tq4.rename(columns={0: "Hits"})
```

```python
Block_test = test[(test.event == "Blocked Shot")]
tq5 = Block_test.groupby(["game_id","team_id_for"]).size()
tq5 = pd.DataFrame(tq5)
tq5 = tq5.reset_index()
tq5 = tq5.rename(columns={0: "Blocked Shots"})
```

```python
Give_test = test[(test.event == "Giveaway")]
tq6 = Give_test.groupby(["game_id","team_id_for"]).size()
tq6 = pd.DataFrame(tq6)
tq6 = tq6.reset_index()
tq6 = tq6.rename(columns={0: "Giveaways"})
```

```python
Miss_test = test[(test.event == "Missed Shot")]
tq7 = Miss_test.groupby(["game_id","team_id_for"]).size()
tq7 = pd.DataFrame(tq7)
tq7 = tq7.reset_index()
tq7 = tq7.rename(columns={0: "Missed Shots"})
```

```python
Penality_test = test[(test.event == "Penalty")]
tq8 = Penality_test.groupby(["game_id","team_id_for"]).size()
tq8 = pd.DataFrame(tq8)
tq8 = tq8.reset_index()
tq8 = tq8.rename(columns={0: "Penalities"})
```

```python
Face_test = test[(test.event == "Faceoff")]
tq9 = Face_test.groupby(["game_id","team_id_for"]).size()
tq9 = pd.DataFrame(tq9)
tq9 = tq9.reset_index()
tq9 = tq9.rename(columns={0: "#Won Faceoffs"})
```

```python
# Renaming columns to prepare for left join merge
gts = gts.rename(columns={'team_id': "team_id_for"})
new_df = pd.merge(gts,tq,how='left', left_on=['game_id','team_id_for'], right_on=['game_id','team_id_for'])
new_df = pd.merge(new_df,tq2,how='left', left_on=['game_id','team_id_for'], right_on=['game_id','team_id_for'])
new_df = pd.merge(new_df,tq3,how='left', left_on=['game_id','team_id_for'], right_on=['game_id','team_id_for'])
new_df = pd.merge(new_df,tq4,how='left', left_on=['game_id','team_id_for'], right_on=['game_id','team_id_for'])
new_df = pd.merge(new_df,tq5,how='left', left_on=['game_id','team_id_for'], right_on=['game_id','team_id_for'])
new_df = pd.merge(new_df,tq6,how='left', left_on=['game_id','team_id_for'], right_on=['game_id','team_id_for'])
new_df = pd.merge(new_df,tq7,how='left', left_on=['game_id','team_id_for'], right_on=['game_id','team_id_for'])
new_df = pd.merge(new_df,tq8,how='left', left_on=['game_id','team_id_for'], right_on=['game_id','team_id_for'])
new_df = pd.merge(new_df,tq9,how='left', left_on=['game_id','team_id_for'], right_on=['game_id','team_id_for'])
```

```python
new_df.head(10)
```

```python
# Nan values represent zero so just replace those
new_df[['Shots','Goals','Takeaways','Hits','Blocked Shots','Giveaways','Missed Shots','Penalities',"#Won Faceoffs"]] = new_df[['Shots','Goals','Takeaways','Hits','Blocked Shots','Giveaways','Missed Shots','Penalities',"#Won Faceoffs"]].fillna(0)
```
